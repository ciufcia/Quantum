### 1. Find the state of a single qubit for which the probability of measuring 0 is four times larger than measuring 1

$$
P(0) = |\alpha|^2 \qquad P(1) = |\beta|^2 \\[1em]
4|\beta|^2 + |\beta|^2 = 1 \implies 5|\beta|^2 = 1 \implies |\beta| = \frac{1}{\sqrt{5}} \iff |\alpha| = \frac{2}{\sqrt{5}}
$$
